```python
print("Challenge 4: String Compressor \n\nImplement a method to perform basic string compression using the counts of repeated characters. (This is called run-length encoding.) For example, the string aabcccccaaa would become a2b1c5a3. If the “compressed” string would not become smaller than the original string, your method should return the original string. You can assume the string has only uppercase and lowercase letters (a–z). Specify whether your solution is case sensitive or case insensitive and what you would need to change to make it the other. Afterward, write a brief explanation walking through your code's logic in markdown.")

print("\n")

print("\n")

# chal4_string = "aabcccccaaa"
chal4_string = str(input("What string would you like to convert?\n"))


#
#"Specify whether your solution is case sensitive or case insensitive and what you would need to change to make it the other."
#
#chal4_string = chal4_string.upper()
#
#
#


print("chal4_string =", chal4_string)

compressed_string = []
letter_counter = 1
for x in range(len(chal4_string)):
    # What to do if it's the very first letter
    if x == 0:
        # set counter to 1
        letter_counter == 1
        # append letter to compressed_string
        compressed_string.append(str(chal4_string[x]))
    # What to do if it's the same as the previous
    if x >= 1:
        if str(chal4_string[x]) == str(chal4_string[x - 1]):
            # Add 1 to counter
            letter_counter += 1
        # What to do if it's not same as the previous
        if chal4_string[x] != chal4_string[x - 1]:
            # append counter to compressed_string
            compressed_string.append(letter_counter)
            # add letter to compressed_string
            compressed_string.append(chal4_string[x])
            # set counter to 1
            letter_counter = 1
# For very last letter, append counter to compressed_string
compressed_string.append(letter_counter)


print("\n")
print("most recent value of letter_counter = ", letter_counter)
print("compressed_string = ", compressed_string)
print("\n")
joined_compressed_string = ""

for x in range(len(compressed_string)):
    joined_compressed_string = joined_compressed_string + str(compressed_string[x])
# "If the “compressed” string would not become smaller than the original string, your method should return the original string."

print("joined_compressed_string = ", joined_compressed_string)
print("len(joined_compressed_string) = ", len(joined_compressed_string))
print("chal4_string =", chal4_string)
print("len(chal4_string) =", len(chal4_string))
final_output = ""
if len(joined_compressed_string) < len(chal4_string):
    final_output = joined_compressed_string
else:
    final_output = chal4_string
print("\n")
print("final_output = ",final_output)
```
